**PROBLEMA DE NEGOCIO**

En un hospital se tiene una alta tasa de reingresos de los pacientes, lo que ha generado que el costo sea significativamente mayor en los pacientes que reingresaron que en los que no reingresaron, además se ha observado que la proporción de mortalidad intrahospitalaria es menor en los pacientes que no han presentado reingresos.Por otro lado debido a los reingresos el hospital en ocasiones se ha quedado corto en capacidad para brindar atención a todos los pacientes lo cual ha generado que varios de los pacientes tengan al hospital en mala imágen.Para atender a este problema la gerencia ha pedido al área de analitica la elaboración de un modelo que prediga los reingresos de los pacientes con el objetivo de planear de forma eficiente su capacidad y mejorar la atención.

**Impacto:**
- Capacidad de atención. 
- Satisfacción del cliente. 
- Nivel de costos.

**DISEÑO DE LA SOLUCIÓN**


**¿Qué se espera realizar con el modelo?**
1. Predecir los pacientes que van a reingresar. 
2. Determinar las variables que influyen en el reingreso.
3. En base a los resultados obtenidos plantear estrategias para prevenir el reingreso de los pacientes y mejorar su atención. 


**Zoom Analítico**
- Los registros de los pacientes de reingreso se actualizaran cada vez que haya un reingreso.
- La base de datos de los demas pacientes se actualizara mensualmente. 
- Semestralmente se va realizar el entrenamiento del modelo para contar con suficientes datos y evaluar su desempeño.
